## Statistical Violence Information in Rio de Janeiro




The statistical overview it's very important to seek if we have some information that are damaged our analysis (outliers), and to see the beginning of ours analysis and what we can extract from it.

```
df.describe()
```

One interesting analise it's teh percent of recovered vehicle in the total that was lost (theft/steal), in this analise we can view the effectiveness of the police department.

  * $\frac{\text{recovered vehicles}}{\text{vehicles theft} + \text{vehicles steal}}$

```
roubo_veiculos = df.roubo_veiculo.mean()
furto_veiculos = df.furto_veiculos.mean()
recuperacao_veiculos = df.recuperacao_veiculos.mean()
print(roubo_veiculos)
print(furto_veiculos)
print(recuperacao_veiculos)
```

```
carros_recuperados = recuperacao_veiculos / (roubo_veiculos + furto_veiculos)
carros_recuperados
```

See how it's the maximum and minimum numbers of murder we have per mounth it's important to see how the violence vary during the time, we will plot a graphic in the next topic to have a better insight.

```
dol_min = df.hom_doloso.min()
dol_max = df.hom_doloso.max()
print(dol_min)
print(dol_max)
```
